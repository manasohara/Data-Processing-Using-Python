```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('height_weight.csv')
fig, ax = plt.subplots(figsize=(15, 5))
#fig-figure
ax.scatter(x = df['height'], y = df['female'])
plt.xlabel("Height of Female")
plt.ylabel("Weight for Female")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('marchcorona.csv')
fig, ax = plt.subplots(figsize=(15, 5))
#fig-figure
ax.scatter(x = df['Cases'], y = df['Day'])
plt.xlabel("Cases")
plt.ylabel("Day")
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

from IPython import get_ipython

get_ipython().run_line_magic('matplotlib', 'inline')

df = pd.read_csv("bank.csv")
fig = plt.figure(figsize =(10, 10))
df.boxplot(by ='Account_Type',grid='True',column =['Balance'], color='red')
```
