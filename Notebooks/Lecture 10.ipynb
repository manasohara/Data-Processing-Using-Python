```python
from scipy import misc
import numpy as np
import matplotlib.pyplot as plt
#getting image from panda and flipping
panda=misc.face()
flip_down=np.flipud(panda)
plt.imshow(flip_down)
plt.show()
```

## mean

```python
import pandas as pd
df=pd.read_csv('product.csv')
x=df["price"].mean()
df["price"].fillna(x,inplace=True)
print(df)
```

## median

```python
import pandas as pd
df=pd.read_csv('product.csv')
x=df["price"].median()
df["price"].fillna(x,inplace=True)
print(df)
```

## loop through all values in price if val>200 set to =350

```python
import pandas as pd
df=pd.read_csv('product.csv')
for x in df.index:
    if df.loc[x,"price"] > 200:
        df.loc[x,"price"]=350
print(df)
```

## drop 

```python
import pandas as pd
df=pd.read_csv('product.csv')
for x in df.index:
    if df.loc[x,"price"] > 200:
        df.drop(x,inplace=True)
print(df)
```
