```python
##a)Use bank.csv file. Plot Sr.No.(SL) against balance. Change dot to some other character e.g. *
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('bank.csv')
fig,ax=plt.subplots(figsize=(10,5))
ax.scatter(x=df['SL'],y=df['Balance'],marker='*')
plt.xlabel("SL")
plt.ylabel("Balance")
plt.show()
```

```python
#b)Change size of all the dots to some specific one e.g. 300/400/500.
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('bank.csv')
fig,ax=plt.subplots(figsize=(12,7))
ax.scatter(x=df['SL'],y=df['Balance'],marker='*',s=500)
plt.xlabel("SL")
plt.ylabel("Balance")
plt.show()
```

```python
##a) Read diamonds.csv file, plot a scatter chart based on caret and depth.
import pandas as pd
import matplotlib.pyplot as plt
df1=pd.read_csv('diamonds.csv')
fig,ax=plt.subplots(figsize=(12,7))
ax.scatter(x=df1['carat'],y=df1['depth'])
plt.xlabel("caret")
plt.ylabel("depth")
plt.show()
```

```python
##b)Plot a scatter chart based on caret and depth.Use different colors for different cut.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df1=pd.read_csv('diamonds.csv')
fig,ax=plt.subplots(figsize=(12,7))
sns.scatterplot(x=df1['carat'],y=df1['depth'],hue=df1['cut'])
plt.xlabel("caret")
plt.ylabel("depth")
plt.show()
```
