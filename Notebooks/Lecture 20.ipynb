```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
y=np.array([35,25,25,15])
subjectlabel=["English","Maths","Science","Geography"]
plt.pie(y,labels=subjectlabel)
style.use('grayscale')
print(plt.style.available)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
y = np.array([10,15,25,10,15,20,5])
subjectlabel = ["House Rent","Education for children","Food","Clothes","Savings","Others","Transport"]
plt.pie(y, labels = subjectlabel)
style.use('grayscale')
labels= ['House Rent', 'Education for children', 'Food', 'Clothes','Savings','Others','Transport']
colors=['blue', 'yellow', 'green', 'orange','black','pink','blue']
sizes= [1500, 600, 500, 300]
plt.pie(sizes,labels=labels, colors=colors, startangle=90, autopct='%1.1f%%')
plt.axis('equal')
#plt.legend(title="Expenditure %")
plt.show() 
```

## Python time series

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('goldsilver.csv')
df = pd.DataFrame(data, columns = ['Date', 'Gold22k'])
# Set the Date as Index
df['Date'] = pd.to_datetime(df['Date'])
df.index = df['Date']
del df['Date']
df.plot(figsize=(8, 6))
plt.show()
```
