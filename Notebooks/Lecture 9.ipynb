```python
# Drawing Pie Chart with label and egend and weiged 0.2 news
import matplotlib.pyplot as plt
import numpy as np
y = np.array([10,15,25,50])
channel = ["Informative","News","Sports","Entertainment"]
myexplode = [0,0.2,0,0]
plt.pie(y,labels=channel,explode=myexplode)
plt.legend(title="Viewers Watching Different types of channels on T.V")
plt.show()
```

```python
import matplotlib.pyplot as plt

import numpy as np



x = np.array([50,25,15,10])

Categorylabels = ["Entertainment " + str(x[0]) +"%" , "Sports " + str(x[1]) + "%" , "News " + str(x[2]) + "%","Informative " + str(x[2]) + "%"]

mycolors=["green","blue","yellow","orange"]

plt.pie(x,labels= Categorylabels ,explode=[0,0,0.2,0],colors=mycolors)

plt.legend(title = "Viewers")

plt.show()
```
